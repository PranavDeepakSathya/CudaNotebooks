```python
from PyLayouts import Swizzled_layout, Draw
import numpy as np
```

```python
def layout_to_addrs (layout_array, layout_element_width, mem_word_width): 
  N_elems_per_word = mem_word_width//layout_element_width
  return layout_array // N_elems_per_word
```

```python
shape = (8,8)
stride = (1,8)
m_base = 0
b_bits = 0
s_shift = 0
elem_width = 4 
mem_width = 32
layout = Swizzled_layout(shape, stride, m_base, b_bits,s_shift)
D_ = Draw(layout.realized_swizzler, 8, 'row')
D_.show(mode = 'address')
addr_map = layout_to_addrs(layout.realized_swizzler, elem_width, mem_width)
D = Draw(addr_map,8, order = 'row')
D.show(mode = 'bank')
```
